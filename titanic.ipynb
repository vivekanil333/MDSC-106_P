```python
import pandas as pd
import seaborn as sns
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
titanic_df=pd.read_csv('Titanic-Dataset.csv')
print(titanic_df)
```

```python
titanic_df.info()
```

```python
titanic_df.describe()
```

```python
titanic_df = titanic_df.dropna(subset=['Age'])
titanic_df #remove rows with missing age values
```

```python
##sort the dataframe by age
titanic_df=titanic_df.sort_values(by='Age')
titanic_df
```

## line plot

```python
plt.figure(figsize=(10,6))
plt.plot(titanic_df['Age'],marker='o', linestyle='-', color='b')
plt.xlabel('Passenger Index')
plt.ylabel('Age')
plt.title('Age Distribution of Titanic Passengers')
plt.grid(True)  # Add grid lines
plt.tight_layout()  # Adjust layout for better appearance
plt.show()
```

## histogram

```python

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_df, x='Age', bins=20, kde=True, color='b')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Titanic Passengers')
plt.grid(True) 
plt.show()
```

The majority of passengers fall within the age range of approximately 20 to 40 years.
There is a notable peak around the early 20s, which could indicate a higher number of young passengers.

## scatter plot

```python


sns.set(style="whitegrid")
plt.figure(figsize=(10, 6)) 
sns.scatterplot(data=titanic_df, x='Age', y='Fare', color='b')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare for Titanic Passengers')
plt.grid(True)  
plt.tight_layout()  
plt.show()
```

from this we can say that the most passengers paid relatively lower price ,some went for higher prices

## box plot

box plot shows the distribution of ages among different passenger classes on the tianic,each box represents the inter quartile range
wit the median marked as line in the box

```python
plt.figure(figsize=(10, 6)) 
sns.boxplot(data=titanic_df, x='Pclass', y='Age', palette='Set3')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Box Plot of Age by Passenger Class for Titanic Passengers')
plt.grid(True)  
plt.tight_layout()  
plt.show()
```

for class 1 we can see  most of the passengers where slightly older than class 2 and class 3

```python
x=titanic_df['Age']
y=titanic_df['Pclass']
plt.bar(x,y)
plt.show()
```

## stacked barplots

```python
class_survival = titanic_df.groupby(['Pclass', 'Survived']).size().unstack()
sns.set(style='whitegrid')

# Create the stacked bar plot
class_survival.plot(kind='bar', stacked=True, figsize=(8, 6))

# Customize labels and title
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.title('Survival Count by Passenger Class')
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])

# Show the plot
plt.show()
```

 The code groups the data by two variables: Pclass (passenger class) and Survived (survival status). This grouping creates a new DataFrame named class_survival that holds the count of passengers for each combination of passenger class and survival status
