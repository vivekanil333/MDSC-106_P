```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
telcomcust1_df=pd.read_csv('TelcomCustomer-Churn_1.csv')
telcomcust1_df
```

```python
telcomcust2_df=pd.read_csv('TelcomCustomer-Churn_2.csv')
telcomcust2_df
```

## merge two dataframes

```python
merged_df = pd.merge(telcomcust1_df, telcomcust2_df, on='customerID', how='inner')
merged_df
```

```python
telcomcust1_df_columns = telcomcust1_df.columns.tolist()
telcomcust2_df_columns = telcomcust2_df.columns.tolist()

merged_columns = merged_df.columns.tolist()

for column in telcomcust1_df_columns:
    if column not in merged_columns:
        print(f"Column '{column}' from df1 is not in the merged DataFrame.")

for column in telcomcust2_df_columns:
    if column not in merged_columns:
        print(f"Column '{column}' from df2 is not in the merged DataFrame.")

if set(telcomcust1_df_columns + telcomcust2_df_columns) == set(merged_columns):
    print("all the column are incorporate into merged data")

```

```python
col = merged_df.select_dtypes(include=['float', 'int']).columns
for column_name in merged_df:
    if(column_name != 'customerID' and  column_name not in col):
        x = merged_df[column_name].value_counts()
        plt.pie(x, labels=x.index, autopct='%1.1f%%', startangle=140)
        plt.title(f'{column_name}')
        plt.show()
```
